Given 3 sequences :

Seq 1 = 'ATGCATGGTGCGCGA’

Seq 2 = 'ATTTGTGCTCCTGGA’

Seq 3 = 'TATGTGCTCCTGGAT’



Find similarity percentage between the three sequence
Check similarity using Hamming Distance, Levenstein Distance, and Dot Plot

```python
!pip install biopython
import Bio
```

```python
from Bio.Seq import Seq
```

```python
seq1 = Seq('ATGCATGGTGCGCGA')
seq2 = Seq('ATTTGTGCTCCTGGA')
seq3 = Seq('TATGTGCTCCTGGAT')
```

Similarity percentage

```python
from Bio import pairwise2
```

```python
AvB = pairwise2.align.localxx(seq1, seq2, one_alignment_only = True, score_only = True)
BvC = pairwise2.align.localxx(seq2, seq3, one_alignment_only = True, score_only = True)
AvC = pairwise2.align.localxx(seq1, seq3, one_alignment_only = True, score_only = True)
```

```python
print("AvB", AvB/len(seq2) *100)
print("BvC", BvC/len(seq2) *100)
print("AvC", AvC/len(seq2) *100)
```

Hamming distance

```python
# Hamming distance fxn
def hamming_distance(lhs, rhs):
  return len([(x,y) for x,y in zip(lhs, rhs) if x != y])
```

```python
print ("Hamming Seq1 vs Seq2 =", hamming_distance(seq1, seq2))
print ("Hamming Seq2 vs Seq3 =", hamming_distance(seq2, seq3))
print ("Hamming Seq1 vs Seq3 =", hamming_distance(seq1, seq3))
```

Levenshtein distance

```python
!pip install python-Levenshtein
from Levenshtein import distance
```

```python
print ("Levenshtein Seq1 vs Seq2 = ", distance (str(seq1), str(seq2)))
print ("Levenshtein Seq2 vs Seq3 = ", distance (str(seq2), str(seq3)))
print ("Levenshtein Seq1 vs Seq3 = ", distance (str(seq1), str(seq3)))
```

Dot plot

seq1 & seq2

```python
def delta(x, y):
  return 0 if x == y else 1

def M(seq1, seq2, i, j, k):
  return sum(delta(x,y) for x,y in zip(seq1[i:i+k], seq2[j:j+k]))

def makeMatrix(seq1, seq2, k):
  n = len(seq1)
  m = len(seq2)
  return [[M(seq1, seq2, i, j, k) for j in range (m-k+1)] for i in range(n-k+1)]

def plotMatrix(M, t, seq1, seq2, nonblank = chr(0x25A0), blank = ' '):
  print(' |' + seq2)
  print('-'*(2 + len(seq2)))
  for label, row in zip(seq1,M):
    line = ''.join(nonblank if s < t else blank for s in row)
    print(label + '|' + line)

def dotplot(seq1, seq2, k = 1, t = 1):
  M = makeMatrix(seq1, seq2, k)
  plotMatrix (M, t, seq1, seq2)
```

```python
dotplot(seq1, seq2)
```
